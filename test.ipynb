```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv("out.csv")
```

```python
plt.figure()
df.hist(bins=100)
```

```python
def normal_distrib(name):
    # Calculating mean and Stdev of AGW
    df_mean = np.mean(df[name])
    df_std = np.std(df[name])
     
    # Calculating probability density function (PDF)
    pdf = stats.norm.pdf(df[name].sort_values(), df_mean, df_std)
    
    # Drawing a graph
    plt.plot(df[name].sort_values(), pdf)
    plt.xlabel(name, size=12)    
    plt.ylabel("Frequency", size=12)                
    plt.grid(True, alpha=0.3, linestyle="--")
    plt.show()
```

```python
normal_distrib("conf")
normal_distrib("xDiff")
normal_distrib("yDiff")
normal_distrib("widthDiff")
normal_distrib("heightDiff")
```
